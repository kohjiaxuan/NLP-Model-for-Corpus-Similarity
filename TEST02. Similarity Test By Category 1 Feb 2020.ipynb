# Philosophers and social scientists

```python
philos = ["Confucius", "Herodotus", "Socrates", "Plato", "Aristotle", "Avicenna", "Thomas Aquinas", "Ibn Khaldun", "René Descartes", "John Locke", "Adam Smith", "Immanuel Kant", "Mary Wollstonecraft", "Karl Marx", "Friedrich Nietzsche", "Sigmund Freud"]

for i in range(len(philos)):
    for j in range(i+1, len(philos)):
        if i == 0 and j == 1:
            df = pd.DataFrame(pd.DataFrame(np.array(similar(philos[i],philos[j], verbose=0).ans())).transpose())
        else:
            df = pd.concat([df, pd.DataFrame(np.array(similar(philos[i],philos[j], verbose=0).ans())).transpose()], axis=0)
            
df.columns = ['Topic 1', 'Topic 2', 'Probability', 'Similar?']
df.index = range(df.shape[0])
            
df
```

```python
df['Similar?'].value_counts()
```

```python
df['Probability'] = df['Probability'].astype('int')
```

```python
df['Probability'].describe()
```

```python
df[df['Similar?'] == 'No']
```

```python
df = df.sort_values('Probability', ascending=False)
```
