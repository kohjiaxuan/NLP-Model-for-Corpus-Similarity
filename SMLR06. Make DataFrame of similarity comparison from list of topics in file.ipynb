```python
from comparewiki import similar
import pandas as pd
import numpy as np

def get_df(filename):
    store = []
    # Store list of topics from file in an array
    with open(filename,'r') as f:
        currentword = f.readline()
        while currentword.strip() != '':
            store.append(currentword.strip().replace('\n', ''))
            currentword = f.readline()
    print(store)
    
    # Build dataframe based on all permutations of topics
    for i in range(len(store)):
        for j in range(i+1, len(store)):
            if i == 0 and j == 1:
                df = pd.DataFrame(pd.DataFrame(np.array(similar(store[i],store[j], verbose=0).ans())).transpose())
            else:
                df = pd.concat([df, pd.DataFrame(np.array(similar(store[i],store[j], verbose=0).ans())).transpose()], axis=0)

    df.columns = ['Topic 1', 'Topic 2', 'Probability', 'Similar?']
    df.index = range(df.shape[0])
            
    return df
```

```python
# Get all political leader similarity
politics_df = get_df('political_leaders.txt')
politics_df
```

```python
# Sort dataframe from highest similarity probability to lowest
politics_df['Probability'] = politics_df['Probability'].astype('int')
politics_df = politics_df.sort_values('Probability', ascending=False)
```

```python
# Save as CSV
politics_df.index.name = 'Index'
politics_df.to_csv('political_leaders.csv', sep='|')
```
